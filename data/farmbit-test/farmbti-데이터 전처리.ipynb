```python
file_path = "국토교통부_표준지공시지가_20240101.csv"
df_f1 = pd.read_csv(file_path,encoding="utf-8-sig")

# 데이터 확인
print(df_f1.head())
print(df_f1.columns)
```

```python
df_regions = pd.read_csv("지역_대분류_소분류.csv", encoding="utf-8-sig")

# 데이터 확인
print(df_regions.head())
```

```python
df_f1.info()
```

```python

# 📌 1️⃣ 필요한 컬럼만 선택
df_f1_selected = df_f1[["시도명", "시군구명", "공시지가"]].copy()

# 📌 2️⃣ 공시지가 전처리 (쉼표 제거 후 숫자로 변환)
df_f1_selected["공시지가"] = df_f1_selected["공시지가"].astype(str).str.replace(",", "").astype(float)

# 📌 3️⃣ `시도명` + `시군구명` 결합하여 `full_region` 생성
df_f1_selected["full_region"] = df_f1_selected["시도명"].str.strip() + " " + df_f1_selected["시군구명"].str.strip()

# 📌 4️⃣ 같은 `full_region` 내 공시지가 평균 계산
df_f1_grouped = df_f1_selected.groupby("full_region", as_index=False)["공시지가"].mean()

# 📌 5️⃣ `df_regions`에도 `full_region` 컬럼 생성
df_regions["full_region"] = df_regions["대분류"].str.strip() + " " + df_regions["소분류"].str.strip()

# 📌 6️⃣ 병합 수행
df_merged = df_regions.merge(df_f1_grouped, on="full_region", how="left")

# 📌 7️⃣ 결과 확인 (주피터 노트북에서 실행할 경우)
df_merged.isna().sum()

```

#### 2. 실거래가

```python
# ✅ 실거래가 데이터 불러오기
file_path = "토지_실거래가_UTF8.csv"
df_f2 = pd.read_csv(file_path, encoding="utf-8-sig")

# ✅ 필요한 컬럼만 선택
df_real_estate = df_f2[['시군구', '지목', '용도지역', '계약년월', '계약면적', '거래금액(만원)']].copy()

# ✅ 시군구에서 '구'가 포함된 경우 시/군만 남기기 (예: '안산시 단원구' → '안산시')
df_real_estate['시군구'] = df_real_estate['시군구'].apply(lambda x: ' '.join(x.split()[:2]) if '구' in x else x)

# ✅ ㎡당 가격 계산 (만원/㎡)
df_real_estate['㎡당 거래가(만원)'] = df_real_estate['거래금액(만원)'].replace(',', '', regex=True).astype(float) / df_real_estate['계약면적'].astype(float)

# ✅ 농림 및 주거 지역만 유지 (대: 대지, 전: 전답, 답: 논)
df_real_estate = df_real_estate[df_real_estate['지목'].isin(['대', '전', '답'])]

# ✅ 시군구별 평균 ㎡당 거래가 계산
df_real_estate = df_real_estate.groupby('시군구', as_index=False)['㎡당 거래가(만원)'].mean()

# ✅ df_merged와 병합
df_merged = df_merged.merge(df_real_estate, left_on="full_region", right_on="시군구", how="left")
```

```python
df_merged
```

```python
df_merged.isna().sum()
```

```python
# ✅ 공시지가 결측값 처리 (같은 시/도 평균으로 대체)
df_merged['공시지가'] = df_merged.groupby('대분류')['공시지가'].transform(lambda x: x.fillna(x.mean()))

# ✅ 실거래가 결측값 처리 (같은 시/도 평균으로 대체)
df_merged['㎡당 거래가(만원)'] = df_merged.groupby('대분류')['㎡당 거래가(만원)'].transform(lambda x: x.fillna(x.mean()))

# ✅ 남은 결측값이 있다면 전국 평균으로 채우기
df_merged['공시지가'].fillna(df_merged['공시지가'].mean(), inplace=True)
df_merged['㎡당 거래가(만원)'].fillna(df_merged['㎡당 거래가(만원)'].mean(), inplace=True)

# ✅ 결측값 확인 (다시 체크)
print(df_merged.isna().sum())
```

### 정규화 및 로그 변환

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# 📂 데이터 불러오기
file_path = "FARM_Standardized_Data.csv"  # 실제 파일 경로로 수정
df = pd.read_csv(file_path)

# ✅ 로그 변환 대상 컬럼 선정 (왜도 기준)
log_transform_cols = ['공시지가', '㎡당 거래가(만원)', '면적당 학교 수', '면적당 병원 수', '인구밀도 (명/㎢)']

# 🚀 로그 변환 수행
for col in log_transform_cols:
    # 값이 0 이하인 경우를 방지하기 위해 최소값 보정
    min_value = df[col].min()
    if min_value <= 0:
        df[col] = df[col] + abs(min_value) + 1
    df[col] = np.log1p(df[col])  # 로그 변환

# ✅ Min-Max 정규화
scaler = MinMaxScaler()
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

# 🚀 변환 후 왜도와 첨도 확인
skewness = df[numeric_cols].skew()
kurtosis = df[numeric_cols].kurtosis()

print("📈 왜도 (Skewness):")
print(skewness)
print("\n🏔️ 첨도 (Kurtosis):")
print(kurtosis)

# 💾 정규화된 데이터 저장
df.to_csv("FARM_normalized.csv", index=False, encoding="utf-8-sig")
print("✅ 정규화 데이터 저장 완료: FARM_normalized.csv")
```

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.linalg import Vectors
from pyspark.ml.classification import KNN

# Spark 세션 시작
spark = SparkSession.builder.appName("FARM Recommendation").getOrCreate()

# CSV 파일 불러오기
file_path = "FARM_normalized.csv"
df = spark.read.csv(file_path, header=True, inferSchema=True)

# 사용자 선호도 입력
user_preferences = {
    "F": 0.7,  # 비용 선호도 (0 ~ 1)
    "A": 0.5,  # 접근성 선호도 (0 ~ 1)
    "R": 0.3,  # 커뮤니티 활성화 선호도 (0 ~ 1)
    "M": 0.8   # 자연환경 선호도 (0 ~ 1)
}

# 사용자 벡터 생성
user_vector = Vectors.dense([
    user_preferences["F"],
    user_preferences["A"],
    user_preferences["R"],
    user_preferences["M"]
])

# 벡터화
assembler = VectorAssembler(
    inputCols=["공시지가", "㎡당 거래가(만원)", "소비자물가지수", "면적당 학교 수", "면적당 병원 수", "인구밀도 (명/㎢)", "커뮤니티 활성화 지수", "대기질 지수", "수질 지수", "녹지 비율"],
    outputCol="features"
)
df = assembler.transform(df)

# KNN 모델 설정
knn = KNN(k=3, featuresCol="features", predictionCol="prediction")

# 모델 학습
model = knn.fit(df)

# 사용자 데이터프레임 생성
user_df = spark.createDataFrame([(user_vector,)], ["features"])

# 추천 결과 예측
predictions = model.transform(user_df)
predictions.show()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, udf
from pyspark.sql.types import FloatType
import numpy as np

# Spark 세션 시작
spark = SparkSession.builder.appName("FARM Recommendation").getOrCreate()

# CSV 파일 불러오기
file_path = "C:\Users\SSAFY\farmbtiFARM_normalized.csv"
df = spark.read.csv(file_path, header=True, inferSchema=True)

# 사용자 선호도 입력
user_preferences = {
    "F": 0.7,  # 비용 선호도 (0 ~ 1)
    "A": 0.5,  # 접근성 선호도 (0 ~ 1)
    "R": 0.3,  # 커뮤니티 활성화 선호도 (0 ~ 1)
    "M": 0.8   # 자연환경 선호도 (0 ~ 1)
}

# 사용자 벡터
user_vector = np.array([user_preferences["F"], user_preferences["A"], user_preferences["R"], user_preferences["M"]])

# 유사도 계산 함수 (코사인 유사도)
def cosine_similarity(arr1, arr2):
    dot_product = np.dot(arr1, arr2)
    norm1 = np.linalg.norm(arr1)
    norm2 = np.linalg.norm(arr2)
    return float(dot_product / (norm1 * norm2))

# UDF로 유사도 계산 함수 등록
@udf(FloatType())
def calculate_similarity(*features):
    data_vector = np.array(features)
    return cosine_similarity(data_vector, user_vector)

# 벡터화
features = ["공시지가", "㎡당 거래가(만원)", "소비자물가지수", "면적당 학교 수", "면적당 병원 수", "인구밀도 (명/㎢)", "커뮤니티 활성화 지수", "대기질 지수", "수질 지수", "녹지 비율"]
df = df.withColumn("similarity", calculate_similarity(*features))

# 유사도 기반 추천 (상위 3개 지역)
recommended = df.orderBy(col("similarity").desc()).limit(3)
recommended.show()
```
